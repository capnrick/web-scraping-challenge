```python
from bs4 import BeautifulSoup as bs
import requests
import os
import pymongo 
from webdriver_manager.chrome import ChromeDriverManager
from splinter import Browser
```

```python
#initialize pyMongo to work with Mongo DBs
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

```python
#define database and collection  
db =client.mars_db
collection = db.items
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
#using beatifulsoup html pasrser to create soup object

url= 'https://redplanetscience.com/'
browser.visit(url)

html = browser.html
soup = bs(html, 'html.parser')

#  example to find all items with these tags   quotes = soup.find_all('span', class_='text')
```

```python
#getting view and verifying the soup object returned html
print(soup.prettify())
```

```python
# saving object that contains html referencing the headlines and short article summary into headlines variable
headlines = soup.find('div', class_="list_text")
headlines
```

```python
#extracting the title and saving into variable
news_title=headlines.find('div', class_='content_title').text
news_title
```

```python
#extracting the text related to the article title summary
news_p=headlines.find('div', class_="article_teaser_body").text
news_p
```
